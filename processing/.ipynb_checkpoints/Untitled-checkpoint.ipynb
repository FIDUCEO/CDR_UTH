```python
from processing import CDRs
from processing import plots
import glob
from os.path import join
import pickle
import numpy as np
import matplotlib.pyplot as plt
import typhon
```

```python
CDR_path = '/scratch/uni/u237/users/tlang/UTH/CDRs/processing'
```

```python
cdrs = {}
files = glob.glob(join(CDR_path, '*NOAA18.pkl'))
```

```python
for f in files:
    with open(f, 'rb') as input:
        c = pickle.load(input)
        cdrs[c.time_coverage_start.month] = c

branches = ['ascending', 'descending']
```

# Brightness Temperature

```python
for b in branches:
    fig, ax = plt.subplots(int(np.ceil(len(cdrs) / 2)), 2, figsize=(15, 10))
    ax = ax.ravel()
    fig.suptitle(b)
    for i in range(len(cdrs)):
        im = plots.plotTb(cdrs[i+1], b, ax=ax[i], vmin=240, vmax=270)
        ax[i].set_title(i+1)
        ax[i].set_xlabel('Longitude')
        ax[i].set_ylabel('Latitude')
    fig.subplots_adjust(right=0.8)
    cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
    fig.colorbar(im, cax=cbar_ax)
    
```

# Observation counts

```python
for b in branches:
    fig, ax = plt.subplots(int(np.ceil(len(cdrs) / 2)), 2, figsize=(15, 10))
    ax = ax.ravel()
    fig.suptitle(b)
    for i in range(len(cdrs)):
        im = plots.plotCount(cdrs[i+1], b, ax=ax[i], vmin=50, vmax=350)
        ax[i].set_title(i+1)
        ax[i].set_xlabel('Longitude')
        ax[i].set_ylabel('Latitude')
    fig.subplots_adjust(right=0.8)
    cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
    fig.colorbar(im, cax=cbar_ax)
```

```python
bins = np.arange(50, 350)
for b in branches:
    fig, ax = plt.subplots(int(np.ceil(len(cdrs) / 2)), 2, figsize=(15, 10))
    ax = ax.ravel()
    fig.suptitle(b)
    for i in range(len(cdrs)):
        ax[i].hist(cdrs[i+1].observation_count_all[b].ravel(), bins=bins, alpha=0.5)
        ax[i].hist(cdrs[i+1].observation_count[b].ravel(), bins=bins, alpha=0.5)
        ax[i].set_title(i+1)
```

# Few observations

```python
numcells = len(cdrs[1].lon) * len(cdrs[1].lat)
percentage = {i: {} for i in cdrs.keys()}
for b in branches:
    fig, ax = plt.subplots(int(np.ceil(len(cdrs) / 2)), 2, figsize=(15, 10))
    ax = ax.ravel()
    fig.suptitle(b)
    for i in range(len(cdrs)):
        lon, lat = np.meshgrid(cdrs[i+1].lon, cdrs[i+1].lat)
        few_overpasses = cdrs[i+1].observation_count[b] <= 150
        percentage[i+1][b] = np.sum(few_overpasses) / numcells
        im = ax[i].pcolormesh(lon, lat, few_overpasses)
        ax[i].set_title(i+1)
        ax[i].set_xlabel('Longitude')
        ax[i].set_ylabel('Latitude')

for i in range(len(cdrs)):
    print('Month {}: Ascending: {:.2f} %, descending: {:.2f} % of grid cells with few observations'.format(
        i+1, percentage[i+1]['ascending'] * 100, percentage[i+1]['descending'] * 100))
```

# Overpass Count

```python
for b in branches:
    fig, ax = plt.subplots(int(np.ceil(len(cdrs) / 2)), 2, figsize=(15, 10))
    ax = ax.ravel()
    fig.suptitle(b)
    for i in range(len(cdrs)):
        lon, lat = np.meshgrid(cdrs[i+1].lon, cdrs[i+1].lat)
        im = ax[i].pcolormesh(lon, lat, cdrs[i+1].observation_count_overpasses[b],
                              cmap=plt.get_cmap('density', 20), vmin=0, vmax=13)
        ax[i].set_title(i+1)
        ax[i].set_xlabel('Longitude')
        ax[i].set_ylabel('Latitude')
    fig.subplots_adjust(right=0.8)
    cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
    fig.colorbar(im, cax=cbar_ax)
    
```

```python
bins = np.arange(0, 15)
for b in branches:
    fig, ax = plt.subplots(int(np.ceil(len(cdrs) / 2)), 2, figsize=(15, 10))
    ax = ax.ravel()
    fig.suptitle(b)
    for i in range(len(cdrs)):
        ax[i].hist(cdrs[i+1].observation_count_overpasses[b].ravel(), bins=bins)
        ax[i].set_title(i+1)
```

# Few Overpasses

```python
numcells = len(cdrs[1].lon) * len(cdrs[1].lat)
percentage = {i: {} for i in cdrs.keys()}
for b in branches:
    fig, ax = plt.subplots(int(np.ceil(len(cdrs) / 2)), 2, figsize=(15, 10))
    ax = ax.ravel()
    fig.suptitle(b)
    for i in range(len(cdrs)):
        lon, lat = np.meshgrid(cdrs[i+1].lon, cdrs[i+1].lat)
        few_overpasses = cdrs[i+1].observation_count_overpasses[b] < 6
        percentage[i+1][b] = np.sum(few_overpasses) / numcells
        im = ax[i].pcolormesh(lon, lat, few_overpasses)
        ax[i].set_title(i+1)
        ax[i].set_xlabel('Longitude')
        ax[i].set_ylabel('Latitude')

for i in range(len(cdrs)):
    print('Month {}: Ascending: {:.2f} %, descending: {:.2f} % of grid cells with few overpasses'.format(
        i+1, percentage[i+1]['ascending'] * 100, percentage[i+1]['descending'] * 100))
```
